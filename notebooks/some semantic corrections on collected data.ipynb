```python
import pandas as pd
from tqdm import trange

df = pd.read_csv("phonesdata.csv")
```

```python
df
```

```python

for i in trange(len(df)):
    df.loc[i, 'specs'] = df.loc[i, 'specs'].replace("Screen", "Screen:")
    df.loc[i, 'specs'] = df.loc[i, 'specs'].replace("pixels", "pixels,")
    df.loc[i, 'specs'] = df.loc[i, 'specs'].replace("Front Camera", "Front Camera:")
    df.loc[i, 'specs'] = df.loc[i, 'specs'].replace("MP", "MP,")
    df.loc[i, 'specs'] = df.loc[i, 'specs'].replace("OS", "OS:")
    df.loc[i, 'specs'] = df.loc[i, 'specs'].replace("Storage", ",Storage:")
    df.loc[i, 'specs'] = df.loc[i, 'specs'].replace("RAM", ",RAM:")
    df.loc[i, 'specs'] = df.loc[i, 'specs'].replace("Battery", ",Battery:")
```

```python
df
```

```python
df.columns = ['Phone Title', 'Specs', 'Price(Kshs)', 'Rating', 'Specs Score', 'Likes']
```

```python
df.head(1)
```

```python
# stripping
df['Phone Title'] = df['Phone Title'].apply(lambda s: str(s).strip())
df.Specs = df.Specs.apply(lambda s: s.strip())
df['Price(Kshs)'] = df['Price(Kshs)'].apply(lambda s: str(s).strip())
df['Rating'] = df['Rating'].apply(lambda s: str(s).strip())
df['Specs Score'] = df['Specs Score'].apply(lambda s: str(s).strip())
df.Likes = df.Likes.apply(lambda s: str(s).strip())
```

```python
df.head()
```

```python
df.to_csv("phonesdata.csv", index=False)
```
